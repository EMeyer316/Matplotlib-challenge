```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
MDD_df = pd.read_csv(mouse_drug_data_to_load)
CTD_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_data_df = pd.merge(MDD_df, CTD_df, how='outer', on='Mouse ID')


# Display the data table for preview
combined_data_df
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
MTV_df = combined_data_df.copy()
del MTV_df['Mouse ID']
del MTV_df['Metastatic Sites']

# Convert to DataFrame
Tumor_Volume_Groups = MTV_df.groupby(['Drug','Timepoint'])

Mean_Tumor_Data_df = Tumor_Volume_Groups.mean()

# Preview DataFrame
Mean_Tumor_Data_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
SEV_df = MTV_df.copy()

# Convert to DataFrame
Tumor_SE_Groups = SEV_df.groupby(['Drug','Timepoint'])

Standard_Error_df = Tumor_SE_Groups.sem()

# Preview DataFrame
Standard_Error_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
Mean_Tumor_Data_df = Mean_Tumor_Data_df.reset_index()
Standard_Error_df = Standard_Error_df.reset_index()
# Preview that Reformatting worked
Mean_Table = pd.pivot_table(Mean_Tumor_Data_df, values=['Tumor Volume (mm3)'], index=['Timepoint'], columns=['Drug'])
SE_Table = pd.pivot_table(Standard_Error_df, values=['Tumor Volume (mm3)'], index=['Timepoint'], columns=['Drug'])

Mean_Table.columns = Mean_Table.columns.droplevel(0)
SE_Table.columns = SE_Table.columns.droplevel(0)

Mean_Table
```

```python
# Generate the Plot (with Error Bars)
means_C = Mean_Table['Capomulin']
sem_C = SE_Table['Capomulin']
means_I = Mean_Table['Infubinol']
sem_I = SE_Table['Infubinol']
means_K = Mean_Table['Ketapril']
sem_K = SE_Table['Ketapril']
means_P = Mean_Table['Placebo']
sem_P = SE_Table['Placebo']
X_Axis_Numbers = Mean_Table.index.values
X_Axis_Max = X_Axis_Numbers.max()


fig, ax = plt.subplots()
ax.errorbar(X_Axis_Numbers, means_C, yerr=sem_C, fmt="o", color = "r", alpha=0.5,linestyle='solid', label="Capomulin")
ax.errorbar(X_Axis_Numbers, means_I, yerr=sem_I, fmt="^", color = "b", alpha=0.5,linestyle='solid', label="Infubinol")
ax.errorbar(X_Axis_Numbers, means_K, yerr=sem_K, fmt="s", color = "g", alpha=0.5,linestyle='solid', label="Ketapril")
ax.errorbar(X_Axis_Numbers, means_P, yerr=sem_P, fmt="d", color = "k", alpha=0.5, linestyle='solid', label="Placebo")
ax.set_xlim(-0.5, X_Axis_Numbers.max()+5)
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Tumor Volume (mm3)")
ax.yaxis.grid(linestyle='-', linewidth='0.25', color='k')
plt.title('Tumor Response to Treatment')
plt.legend(loc="best", fontsize="small", fancybox=True)
# Save the Figure
plt.savefig('TumorTreatment.png')
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
MMS_df = combined_data_df.copy()
del MMS_df['Mouse ID']
del MMS_df['Tumor Volume (mm3)']

# Convert to DataFrame
MetSite_Groups = MMS_df.groupby(['Drug','Timepoint'])

Mean_MetSite_Data_df = MetSite_Groups.mean()

# Preview DataFrame
Mean_MetSite_Data_df
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
SEMS_df = MMS_df.copy()

# Convert to DataFrame
MS_SE_Groups = SEMS_df.groupby(['Drug','Timepoint'])

Standard_Error_MS_df = MS_SE_Groups.sem()

# Preview DataFrame
Standard_Error_MS_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
Mean_MetSite_Data_df = Mean_MetSite_Data_df.reset_index()
Standard_Error_MS_df = Standard_Error_MS_df.reset_index()
# Preview that Reformatting worked
Mean_MS_Table = pd.pivot_table(Mean_MetSite_Data_df, values=['Metastatic Sites'], index=['Timepoint'], columns=['Drug'])
SE_MS_Table = pd.pivot_table(Standard_Error_MS_df, values=['Metastatic Sites'], index=['Timepoint'], columns=['Drug'])

Mean_MS_Table.columns = Mean_MS_Table.columns.droplevel(0)
SE_MS_Table.columns = SE_MS_Table.columns.droplevel(0)

Mean_MS_Table
```

```python
# Generate the Plot (with Error Bars)
means_C = Mean_MS_Table['Capomulin']
sem_C = SE_MS_Table['Capomulin']
means_I = Mean_MS_Table['Infubinol']
sem_I = SE_MS_Table['Infubinol']
means_K = Mean_MS_Table['Ketapril']
sem_K = SE_MS_Table['Ketapril']
means_P = Mean_MS_Table['Placebo']
sem_P = SE_MS_Table['Placebo']
X_Axis_Numbers = Mean_MS_Table.index.values
X_Axis_Max = X_Axis_Numbers.max()


fig, ax = plt.subplots()
ax.errorbar(X_Axis_Numbers, means_C, yerr=sem_C, fmt="o", color = "r", alpha=0.5,linestyle='solid', label="Capomulin")
ax.errorbar(X_Axis_Numbers, means_I, yerr=sem_I, fmt="^", color = "b", alpha=0.5,linestyle='solid', label="Infubinol")
ax.errorbar(X_Axis_Numbers, means_K, yerr=sem_K, fmt="s", color = "g", alpha=0.5,linestyle='solid', label="Ketapril")
ax.errorbar(X_Axis_Numbers, means_P, yerr=sem_P, fmt="d", color = "k", alpha=0.5, linestyle='solid', label="Placebo")
ax.set_xlim(-0.5, X_Axis_Numbers.max()+5)
ax.set_xlabel("Treatment Duration (Days)")
ax.set_ylabel("Metastatic Sites")
ax.yaxis.grid(linestyle='-', linewidth='0.25', color='k')
plt.title('Metastatic Spread During Treatment')
plt.legend(loc="best", fontsize="small", fancybox=True)
# Save the Figure
plt.savefig('MetastaticTreatment.png')
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
SR_df = combined_data_df.copy()
del SR_df['Metastatic Sites']
del SR_df['Tumor Volume (mm3)']

# Convert to DataFrame
Survival_Groups = SR_df.groupby(['Drug','Timepoint'])

Survival_Rates = Survival_Groups['Mouse ID'].nunique()
Survival_Rates_df = pd.DataFrame(Survival_Rates)
Survival_Rates_df.rename(columns = {'Mouse ID':'Live Mice Count'}, inplace = True) 
Survival_Rates_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
Survival_Count_Data_df = Survival_Rates_df.reset_index()

# Preview that Reformatting worked
Surviving_Mice_Table = pd.pivot_table(Survival_Count_Data_df, values=['Live Mice Count'], index=['Timepoint'], columns=['Drug'])

Surviving_Mice_Table.columns = Surviving_Mice_Table.columns.droplevel(0)

Surviving_Mice_Table
(Surviving_Mice_Table.get_value(0, 'Capomulin', takeable=False))
```

```python
# Generate the Plot (Accounting for percentages)
LiveMice_C = Surviving_Mice_Table['Capomulin']
LiveMice_Percent_C = LiveMice_C / (Surviving_Mice_Table.get_value(0, 'Capomulin', takeable=False)) * 100
LiveMice_I = Surviving_Mice_Table['Infubinol']
LiveMice_Percent_I = LiveMice_I / (Surviving_Mice_Table.get_value(0, 'Infubinol', takeable=False)) * 100
LiveMice_K = Surviving_Mice_Table['Ketapril']
LiveMice_Percent_K = LiveMice_K / (Surviving_Mice_Table.get_value(0, 'Ketapril', takeable=False)) * 100
LiveMice_P = Surviving_Mice_Table['Placebo']
LiveMice_Percent_P = LiveMice_P / (Surviving_Mice_Table.get_value(0, 'Capomulin', takeable=False)) * 100
X_Axis_Numbers = Mean_MS_Table.index.values
X_Axis_Max = X_Axis_Numbers.max()


fig, ax = plt.subplots()
ax.errorbar(X_Axis_Numbers, LiveMice_Percent_C, fmt="o", color = "r", alpha=0.5,linestyle='solid', label="Capomulin")
ax.errorbar(X_Axis_Numbers, LiveMice_Percent_I, fmt="^", color = "b", alpha=0.5,linestyle='solid', label="Infubinol")
ax.errorbar(X_Axis_Numbers, LiveMice_Percent_K, fmt="s", color = "g", alpha=0.5,linestyle='solid', label="Ketapril")
ax.errorbar(X_Axis_Numbers, LiveMice_Percent_P, fmt="d", color = "k", alpha=0.5,linestyle='solid', label="Placebo")
ax.set_xlim(-0.5, X_Axis_Numbers.max()+5)
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Survival Rate (%)")
ax.yaxis.grid(linestyle='-', linewidth='0.25', color='k')
plt.title('Survival During Treatment')
plt.legend(loc="best", fontsize="small", fancybox=True)
# Save the Figure
plt.savefig('SurvivalRates.png')
# Show the Figure
plt.show()
```
